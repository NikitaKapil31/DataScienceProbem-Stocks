CAPM(Capital Asset Pricing Model) attempts to price securities by examining the relationship that exists between expected returns and risk. 
Read more about CAPM. (Investopedia CAPM reference) 
The Beta of an asset is a measure of the sensitivity of its returns relative to a market benchmark (usually a market index). How sensitive/insensitive is the returns of an asset to the overall market returns (usually a market index like S&P 500 index). What happens when the market jumps, does the returns of the asset jump accordingly or jump somehow?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('GOLD.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data_lr = data.dropna()
```

```python
data_lr.shape
```

```python
x = data_lr[['Price', 'Open', 'High', 'Low']]
y = data_lr['Pred']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33,random_state = 42)
```

```python
x_train.head()
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
coeffs = pd.DataFrame(lm.coef_, x.columns, columns=['Coefficient'])
coeffs
```

```python
predictions = lm.predict(x_test)
```

```python
data['Pred']=lm.predict(data[['Price','Open','High','Low']])
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
sns.distplot(y_test-predictions,bins=30)
```

```python
import statsmodels.api as sm
X = data_lr[['Price','Open','High','Low']]
y1 = data_lr['Pred']
X1 = sm.add_constant(X)
model = sm.OLS(y1,X1)
results1=model.fit()
print(results1.summary())
```

```python
import statsmodels.api as sm
X = data[['Price','Open','High','Low']]
y = data['new']
X1 = sm.add_constant(X)
model = sm.OLS(y,X1)
results = model.fit()
print(results.summary())
```
